```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/salmon_data/all_chromosomes.csv", index_col=0)
pivoted_df = df.pivot_table(index='sample.id', columns='gene', values=["gene.expression"],  aggfunc='mean')

X = pivoted_df
X.columns = X.columns.droplevel(0)
pivoted_df = df.pivot_table(index='sample.id', values=["gutted.weight.kg"],  aggfunc='mean')
y = pivoted_df
```

```python
X.index
```

```python
X_reformatted = pd.DataFrame(data= X.values, columns=list(X.columns), index = list(X.index))
X_reformatted
```

```python
XY = pd.merge(left=X_reformatted, right=y, left_on=X_reformatted.index, right_on=y.index)
XY.index = list(XY["key_0"])
XY.drop("key_0", axis=1, inplace=True)
```

```python
XY
```

```python
XY.to_csv("../../data/processed/transcriptome_XY.csv")
```
